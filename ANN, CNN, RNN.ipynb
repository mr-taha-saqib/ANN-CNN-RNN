```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras. layers import Dropout
```

```python
#Q1 ANN
```

```python
dataset= pd.read_csv('Churn_Modelling.csv')
dataset.head()
```

```python
X= dataset.iloc[:,3:-1]
y=dataset.iloc[:,-1]
X.head()
```

```python
geography=pd.get_dummies(X["Geography"],drop_first=True)
gender= pd.get_dummies(X['Gender'],drop_first=True)
```

```python
X=pd.concat([X,geography,gender],axis=1)
X.drop(['Geography','Gender'],axis=1,inplace=True)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
classifier= Sequential()
```

```python
# Adding the input layer and the first hidden layer
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=11))
# Adding the second hidden layer
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))

# Adding the output layer
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
classifier.summary()
```

```python
classifier.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#fitting the ANN to the training set
model_history=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=100)
```

```python
y_pred= classifier.predict(X_test)
y_pred= (y_pred>0.5)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

cm= confusion_matrix(y_test,y_pred)
print(cm)
accuracy=accuracy_score(y_test,y_pred)
print('The accuracy of model is',accuracy*100,'%')
cl_report = classification_report(y_test,y_pred)
print(cl_report)
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('reddit-comments-2015-08.csv')
```

```python
sentences = data['body'].values
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
tokenized_sentences = tokenizer.texts_to_sequences(sentences)
word_index = tokenizer.word_index
vocab_size = len(word_index)
```

```python
labels = data['body'].values
max_len = 100  #  maximum sequence length
padded_sequences = pad_sequences(tokenized_sentences, maxlen=max_len)

X_train, X_test, y_train, y_test = train_test_split(padded_sequences, labels, test_size=0.2, random_state=42)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size + 1, 100, input_length=max_len),  # Word Embedding Layer
    tf.keras.layers.SimpleRNN(128),  # RNN layer with 128 units
    tf.keras.layers.Dense(1, activation='sigmoid')  # Output layer
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
word_dim = 100  # Size of word embedding
hidden_dim = 128  # Size of hidden layer
output_dim = 1  # Output dimension
bptt_truncate = 5  # Truncation
```

```python
# Random initialization of weights
U = np.random.uniform(-np.sqrt(1. / word_dim), np.sqrt(1. / word_dim), (hidden_dim, word_dim))
V = np.random.uniform(-np.sqrt(1. / hidden_dim), np.sqrt(1. / hidden_dim), (output_dim, hidden_dim))
W = np.random.uniform(-np.sqrt(1. / hidden_dim), np.sqrt(1. / hidden_dim), (hidden_dim, hidden_dim))
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def forward_propagation(x):
    hidden_states = np.zeros((len(x), hidden_dim))
    outputs = np.zeros((len(x), output_dim))

    for t in range(len(x)):
        # Update hidden state
        hidden_states[t] = sigmoid(np.dot(U, x[t]) + np.dot(W, hidden_states[t - 1]))

        outputs[t] = sigmoid(np.dot(V, hidden_states[t]))

    return outputs, hidden_states
```

```python
def calculate_loss(y, outputs):
    return -np.mean(y * np.log(outputs) + (1 - y) * np.log(1 - outputs))
```
